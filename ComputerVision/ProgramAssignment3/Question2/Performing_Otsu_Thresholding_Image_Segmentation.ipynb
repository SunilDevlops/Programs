# Performing Otsu Thresholding

### Loading the libraries

```python
from matplotlib import image
import matplotlib.pyplot as plt
import numpy as np
```

### Reading the First Image file

```python
I = image.imread("Image1.jpg")
plt.imshow(I,cmap="gray")
plt.show()
```

### Converting input image into an array

```python
img = np.asarray(I)
img
```

### Plotting the histogram

```python
pixel_counts = [np.sum(img == i) for i in range(256)]
x = np.arange(0,256)
plt.bar(x,pixel_counts, color='black', width=5, align='center')
plt.show()
```

### Function to compute Best Threshold

```python
def otsuThresholding(img):
    # Initialization
    arr = (0,0)
    for threshold in range(256):
        # update weights
        weight_background = sum(pixel_counts[:threshold])
        weight_foreground = sum(pixel_counts[threshold:])
        # update mean
        mu_background = sum([i * pixel_counts[i] for i in range(0,threshold)]) / weight_background if weight_background > 0 else 0       
        mu_foreground = sum([i * pixel_counts[i] for i in range(threshold, 256)]) / weight_foreground if weight_foreground > 0 else 0

        # update variance
        inter_class_variance = weight_background * weight_foreground * (mu_background - mu_foreground) ** 2
        
        if inter_class_variance > arr[1]:
            arr = (threshold, inter_class_variance)     
    return arr[0]
```

### Printing the optimal Threshold for Image 1

```python
optimalThreshold = otsuThresholding(img)
optimalThreshold
```

### Redrawing the image with best threshold into consideration

```python
ones = np.ones_like(I)
zeros = np.zeros_like(I)
I_thresholded = np.where(I > optimalThreshold, ones, zeros)
plt.imshow(I_thresholded, cmap="gray")
```

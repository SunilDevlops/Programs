# Performing binarization process

### Loading the libraries

```python
from matplotlib import image
import matplotlib.pyplot as plt
import numpy as np
```

### Reading the First Image file

```python
I = image.imread("Image1.jpg")
plt.imshow(I,cmap="gray")
plt.show()
```

### Displaying the histogram plot out of it

```python
plt.hist(I.flatten(), bins=256, range=(0,256))
plt.show()
```

### Guessing the threshold value by seeing the histogram plot

```python
threshold = 85
ones = np.ones_like(I)
zeros = np.zeros_like(I)
I_thresholded = np.where(I > threshold, ones, zeros)
plt.imshow(I_thresholded, cmap="gray")
```
